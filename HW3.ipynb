```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
```

```python
mu,sigma = 0,np.sqrt(0.2)
```

```python
data = np.random.normal(mu,sigma,1000)
```

```python
plt.figure()
count, bins, ignored = plt.hist(data,100, density=True)
plt.plot(np.linspace(-1.5,1.5,1000),norm.pdf(np.linspace(-1.5,1.5,1000),mu,sigma),linewidth = 2,color='r')
plt.show()
```
